# Notebook: DPLL - Solución del Problema de Fichas

## Introducción al Algoritmo DPLL

El algoritmo DPLL (Davis–Putnam–Logemann–Loveland) es un método eficiente para la comprobación de satisfacibilidad de fórmulas proposicionales en forma normal conjuntiva (CNF).

**Ventajas:**
- Mucho más eficiente que la tabla de verdad para problemas medianos y grandes.
- Permite encontrar modelos o demostrar insatisfacibilidad.

**Limitaciones:**
- Aunque es eficiente, el problema SAT sigue siendo NP-completo.

En este notebook aplicamos DPLL al problema de las fichas (tablero 4x4, 3 figuras).
Usamos la clase `Fichas` de `Archivos/Proyecto.py` para obtener las 7 reglas y la función `tseitin` para convertir las fórmulas a CNF.

```python
# Imports y preparación
import Archivos
from Archivos import Fichas, tseitin, dpll, Logica
import time
p = Fichas()
reglas = p.reglas
print('Número de reglas:', len(reglas))
for i, r in enumerate(reglas, start=1):
    print(f'Regla {i}:', str(r)[:120] + ('' if len(str(r))<120 else '...'))
```

## 1) Pruebas individuales por regla
Para cada regla (1..7) ejecutamos DPLL y visualizamos el modelo encontrado (si existe).
Se registra el tiempo de ejecución.

```python
# Regla 1: al menos un círculo
A1 = reglas[0]
S1 = tseitin(A1)
t0 = time.time()
res1, I1 = dpll(S1, {})
t1 = time.time()
print('Regla 1 - SAT:', res1, 'Tiempo(s):', round(t1-t0,6))
if res1 == 'Satisfacible':
    p.visualizar(I1)
```
