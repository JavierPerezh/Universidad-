## Introducción al Algoritmo DPLL

El algoritmo DPLL (Davis–Putnam–Logemann–Loveland) es un método eficiente para la comprobación de satisfacibilidad de fórmulas proposicionales en forma normal conjuntiva (CNF).

**Ventajas:**
- Mucho más eficiente que la tabla de verdad para problemas medianos y grandes.
- Permite encontrar modelos o demostrar insatisfacibilidad.

**Limitaciones:**
- Aunque es eficiente, el problema SAT sigue siendo NP-completo.

En este notebook aplicamos DPLL al problema de las fichas (tablero 4x4, 3 figuras).
Usamos la clase `Fichas` de `Archivos/Proyecto.py` para obtener las 7 reglas y la función `tseitin` para convertir las fórmulas a CNF.

## Análisis de resultados <a class="anchor" id="analisis"></a>

Fue uno de los métodos que mejor funcionó. A diferencia de los Tableaux, DPLL usa técnicas más inteligentes como la "propagación unitaria", que le permite deducir valores automáticamente sin tener que probar todas las posibilidades. Por ejemplo, si sabe que en una casilla solo puede haber una ficha y ya puso un círculo, automáticamente deduce que no puede haber triángulo ni cuadrado en esa misma casilla. Esta capacidad de deducción hace que sea mucho más rápido que los métodos anteriores.
